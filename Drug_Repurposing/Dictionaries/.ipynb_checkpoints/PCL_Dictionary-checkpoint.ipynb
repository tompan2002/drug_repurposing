# PCL Dictionary

Similar concept to the ATC dictionary except with PCL drug codes. PCL to Broad ID data from: https://api.clue.io/api/pcls?filter={%22include%22:{%22perts%22:true}}&user_key=b604e3145b4aca58ef72312276e295f3

```python
import pandas as pd
import json
```

Initializing lists which will be combined into a dataframe

```python
pcl_codes = []
pert_ids = []
```

Processing raw data

```python
with open("C:/Users/tp36t/Desktop/Research/pcls.json", 'r') as fp:
    data = json.load(fp)
    for i in range(len(data)):
        group = data[i]["perts"]
        for j in range(len(group)):
            if "pcl_membership" in group[j]:
                if group[j]["pcl_membership"]:
                    pcl_codes.append(group[j]["pcl_membership"][0])
                    pert_ids.append(group[j]["pert_id"])
```

Checking if lists populated correctly

```python
pcl_codes
```

```python
pert_ids
```

Creating dictionary

```python
pcl_to_broad_id_dictionary = pd.DataFrame({"pcl_codes" : pcl_codes, "pert_ids": pert_ids})
pcl_to_broad_id_dictionary
```

Importing sig_info metadata (downloaded from https://clue.io/data/CMap2020#LINCS2020)

```python
sig_info = pd.read_csv("siginfo_beta.txt", sep = "\t", dtype=str)
sig_info
```

Creating function which returns sig_ids from PCL code (exact, not prefix)

```python
def pcl_to_records(pcl):
    records = pcl_to_broad_id_dictionary.loc[pcl_to_broad_id_dictionary.pcl_codes == pcl, "pert_ids"]
    sig_ids = sig_info.merge(records, left_on="pert_mfc_id", right_on="pert_ids", how="inner").sig_id
    return sig_ids
```

```python
pcl_to_records("CP_TUBULIN_INHIBITOR")
```
