# Extracting sig_ids

This code extracts the sig_ids from the ATC codes of two drug classes of interest, C02 which antihypersensitives (https://www.whocc.no/atc_ddd_index/?code=C02), and L01, which are antineoplastic agents (https://www.whocc.no/atc_ddd_index/?code=L01).

```python
import pandas as pd
```

Importing ATC to pert_id dictionary created earlier

```python
dictionary = pd.read_csv("../Dictionaries/ATC_to_pert_id_dictionary.txt")
dictionary
```

Importing sig_info metadata

```python
sig_info = pd.read_csv("../Dictionaries/siginfo_beta.txt", sep = "\t", dtype = str)
sig_info
```

Function that returns sig_ids from ATC code prefix (from ATC_Dictionary)

```python
def ATC_to_records(atc_code):
    bool_series = dictionary.ATC_Code.str.startswith(atc_code, na = False)
    records = dictionary[bool_series]
    pert_ids = records.pert_id
    sig_ids = sig_info.merge(pert_ids, on= "pert_id", how = "inner")
    return sig_ids.sig_id
```

Extracting sig_ids for C02

```python
antihypersensitives = ATC_to_records("C02")
antihypersensitives.to_csv("antihypersensitives.csv")
antihypersensitives
```

```python
antineoplastics = ATC_to_records("L01")
antineoplastics.to_csv("antineoplastics.csv")
antineoplastics
```
