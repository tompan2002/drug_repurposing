```python
import pandas as pd
```

## Importing Signature Info Metadata

```python
sig_info = pd.read_csv("GSE70138_Broad_LINCS_sig_info.txt", sep = "\t")
sig_info
```

## Sample / Column Annotations

```python
sig_info = pd.read_csv("GSE70138_Broad_LINCS_sig_info.txt", sep = "\t")
sig_info.columns
```

Selecting sig_ids of signatures generated by treatment with Vorinostat

```python
vorinostat_ids = sig_info.loc[sig_info.pert_iname =="vorinostat" , "sig_id"]
len(vorinostat_ids)
```

## Gene / Row Annotations

Importing gene info metadata

```python
gene_info = pd.read_csv("GSE92742_Broad_LINCS_gene_info.txt", sep = "\t", dtype = str)
gene_info.columns
```

Selecting pr_gene_ids of landmark genes

```python
landmark_gene_row_ids = gene_info.loc[gene_info.pr_is_lm == "1", "pr_gene_id"]
landmark_gene_row_ids
```

## Parsing gctx File

```python
from cmapPy.pandasGEXpress.parse import parse
```

Parsing metadata from .gctx

```python
my_col_metadata = parse("GSE70138_Broad_LINCS_Level5_COMPZ_n118050x12328.gctx", col_meta_only = True)
my_col_metadata
```

Parsing dataset for vorinostat signatures (sample subset)

```python
vorinostat_only_gctoo = parse("GSE70138_Broad_LINCS_Level5_COMPZ_n118050x12328.gctx", cid = vorinostat_ids)
vorinostat_only_gctoo
```

Parsing dataset for landmark gene signatures (gene subset)

```python
landmark_only_gctoo = parse("GSE70138_Broad_LINCS_Level5_COMPZ_n118050x12328.gctx", rid = landmark_gene_row_ids)
landmark_only_gctoo
```

Pulling vorinostat rows from the metadata

```python
vorinostat_sig_id_info = sig_info[sig_info.pert_iname == "vorinostat"]
vorinostat_sig_id_info
```

Resetting index so it can be put into gctoo

```python
vorinostat_sig_id_info.set_index("sig_id", inplace = True)
vorinostat_sig_id_info
```

## Adding / Deleting gctoo Annotations

Adding metadata to dataset

```python
vorinostat_only_gctoo.col_metadata_df = vorinostat_sig_id_info
vorinostat_only_gctoo.col_metadata_df
```

Deleting a column

```python
del vorinostat_only_gctoo.col_metadata_df["pert_type"]
vorinostat_only_gctoo.col_metadata_df
```

## Taking Subset of gctx

```python
import cmapPy.pandasGEXpress.subset_gctoo as sg
```

Find cid values with A375

```python
vorinostat_a375_ids = vorinostat_only_gctoo.col_metadata_df.index[vorinostat_only_gctoo.col_metadata_df["cell_id"] == "A375"].tolist()
vorinostat_a375_ids
```

Subset gctoo for values of interest

```python
vorinostat_a375_gctoo = sg.subset_gctoo(vorinostat_only_gctoo, cid=vorinostat_a375_ids)
vorinostat_a375_gctoo
```

Pulling belinostat a375 sig info

```python
belinostat_a375_sig_info = sig_info[(sig_info["pert_iname"] == "belinostat") & (sig_info["cell_id"] == "A375")]
belinostat_a375_sig_info
```

Resetting index

```python
belinostat_a375_sig_info.set_index("sig_id", inplace=True)
belinostat_a375_sig_info
```

Parsing dataframe for belinostat A375 data

```python
belinostat_a375_gctoo = parse("GSE70138_Broad_LINCS_Level5_COMPZ_n118050x12328.gctx", cid = belinostat_a375_sig_info.index)
belinostat_a375_gctoo
```

Annotating belinostat a375 gctoo with metadata

```python
belinostat_a375_gctoo.col_metadata_df = belinostat_a375_sig_info.copy()
belinostat_a375_gctoo.col_metadata_df
```

## Concatenating gctx Files

```python
import cmapPy.pandasGEXpress.concat as cg
```

Horizontally stack gctoo instances

```python
HDAC_a375_gctoo = cg.hstack([vorinostat_a375_gctoo, belinostat_a375_gctoo])
HDAC_a375_gctoo
```

gctx corresponding to two compounds in one cell line:

```python
print("unique compound data present: ")
print(HDAC_a375_gctoo.col_metadata_df["pert_iname"].unique())
print("unique cell line data present: ")
print(HDAC_a375_gctoo.col_metadata_df["cell_id"].unique())
```

Check same genes as before

```python
print("combined gctoo expression data shape:")
print(HDAC_a375_gctoo.data_df.shape)
```

## Writing gctx Files

```python
import cmapPy.pandasGEXpress.write_gctx as wg
```

Write to gctx file

```python
wg.write(belinostat_a375_gctoo, "belinostat_a375")
```
